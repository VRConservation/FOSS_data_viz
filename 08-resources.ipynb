# 8. Resources
A curated collection of resources, tutorials, and references for geospatial data visualization using free and open source tools.

## Educational Resources & Blogs

1. [Spatial Thoughts](https://spatialthoughts.com/)
Excellent tutorials and courses on geospatial analysis, QGIS, and Python. High-quality content for both beginners and advanced users.
2. [John Nelson - Mapping & Visualization](https://www.esri.com/en-us/professional-services/professional-services-team/john-nelson)
While primarily focused on ArcGIS Pro, John Nelson's mapping tutorials contain valuable principles and techniques applicable to QGIS and other FOSS tools. **Note:** Some specific tools are ArcGIS Pro-specific, but conceptual knowledge transfers well.
3.  [FOS Geospatial Tools](https://3point.xyz/geosp/)
Focused resources on geospatial concepts and practical applications for data visualization.

## Python & Developer Resources

1. [Qiusheng Wu's GitHub (giswqs)](https://github.com/giswqs)
Extensive collection of geospatial Python packages and tools, including:
    - **leafmap** - Interactive mapping library
    - **geemap** - Google Earth Engine Python API wrapper
    - Sample notebooks and tutorials for geospatial analysis
2.  Official Documentation
    - [GeoPandas Documentation](https://geopandas.org/) - Working with geospatial data in Python
    - [Rasterio Documentation](https://rasterio.readthedocs.io/) - Raster data access and processing
    - [GDAL/OGR Documentation](https://gdal.org/) - Geospatial data translation and processing
    - [Folium Documentation](https://python-visualization.github.io/folium/) - Interactive maps with Python
    - [Leafmap Documentation](https://leafmap.org/) - Geospatial analysis and interactive mapping

## FOSS GIS Software

1. QGIS
    - **Official Site:** https://www.qgis.org/
    - [QGIS User Guide](https://docs.qgis.org/)
    - [QGIS Python API (PyQGIS)](https://docs.qgis.org/3.28/en/docs/pyqgis_developer_guide/)
2.  GDAL
    - **Official Site:** https://gdal.org/
    - Command-line tools for raster and vector data processing
    - Powerful for batch processing and automation
3.  PostGIS
    - Spatial extension for PostgreSQL
    - Excellent for large-scale geospatial analysis
    - https://postgis.net/
4. GRASS GIS
    - Powerful raster and vector processing
    - https://grass.osgeo.org/

## Data Visualization Libraries

1. Interactive Maps
    - **Leafmap** - Jupyter integration for interactive maps
    - **Folium** - Leaflet.js wrapper for Python
    - **Plotly** - Interactive charts and maps
    - **Geopandas** - Direct mapping integration
2. Static Visualization
    - **Matplotlib** - Foundational plotting library
    - **Cartopy** - Cartographic projections and mapping
    - **Contextily** - Background map tiles for matplotlib
3. Advanced Visualization
    - **Seaborn** - Statistical data visualization
    - **Holoviews** - Complex multidimensional data
    - **Datashader** - Large dataset visualization

## Data Sources

1. Global Datasets
    - **USGS Earth Explorer** - Landsat, Sentinel imagery (https://earthexplorer.usgs.gov/)
    - **Google Earth Engine** - Petabytes of geospatial data (https://earthengine.google.com/)
    - **OpenStreetMap** - Free geographic data (https://www.openstreetmap.org/)
    - **Natural Earth Data** - Natural and cultural datasets (https://www.naturalearthdata.com/)
    - **Copernicus Climate Data** - EU climate and environmental data (https://cds.climate.copernicus.eu/)
2. Administrative Boundaries
    - **Natural Earth Admin Boundaries** - Countries and subdivisions
    - **GADM** - Global administrative divisions (https://gadm.org/)
    - **UN World Boundaries** - Official UN boundary data
3. Elevation & Terrain
    - **SRTM** - Shuttle Radar Topography Mission data
    - **GEBCO** - General Bathymetric Chart of the Oceans (https://www.gebco.net/)
    - **OpenTopography** - High-resolution topography data (https://www.opentopodata.org/)

## Useful Tools & Utilities

- **GDAL Command Line Tools** - gdalinfo, gdaltranslate, ogr2ogr
- **shp2pgsql** - Load shapefiles into PostGIS
- **tippecanoe** - Build vector tilesets from newline-delimited GeoJSON
- **mapshaper** - Simplify GeoJSON, shapefile, and CSV data (https://mapshaper.org/)
- **geojson.io** - Web-based GeoJSON editor (https://geojson.io/)
- **QGIS Processing Toolbox** - Integrated analysis workflows
